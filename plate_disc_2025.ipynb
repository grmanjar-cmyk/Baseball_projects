```python
import pandas as pd

# Load your corrected data files
df_out_zone = pd.read_csv('yty_change_out_zone_swing_percent.csv')
df_in_zone = pd.read_csv('yty_change_in_zone_swing_percent.csv')
df_hitting = pd.read_csv('2025_hitting_rate_stats.csv')
```

```python
# Merge the two dataframes using the player ID and name
df_merged = pd.merge(
    df_out_zone,
    df_in_zone,
    on=['entity_id', 'name'],
    suffixes=('_out_zone', '_in_zone') #suffixes are added to create columns for in_zone and out_zone swings for each year
)
# First, let's look at the top 5 rows to make sure it worked
print("First 5 rows of the merged data:")
df_merged.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#plot scatter plot comparing in and out of zone swing rates to look for potential correlation

sns.scatterplot(data=df_merged, x='2025_out_zone', y='2025_in_zone')
plt.title('In-Zone Swing % vs. Out-of-Zone Swing % (2025)')
plt.xlabel('Out-of-Zone Swing % (O-Swing%)') # Adding label
plt.ylabel('In-Zone Swing % (Z-Swing%)')   # Adding label
plt.show()
```

```python
# --- Step 1: Calculate the averages ---
# preparing new plot to show plate discipline quadrants
avg_out_zone = df_merged['2025_out_zone'].mean()
avg_in_zone = df_merged['2025_in_zone'].mean()

# --- Create the main scatter plot ---
plt.figure(figsize=(10, 8)) # Make the plot a bit bigger
sns.scatterplot(data=df_merged, x='2025_out_zone', y='2025_in_zone')

# --- Step 2: Draw the quadrant lines ---
# Add a vertical line for the average out-of-zone swing %
plt.axvline(x=avg_out_zone, color='red', linestyle='--', linewidth=1)
# Add a horizontal line for the average in-zone swing %
plt.axhline(y=avg_in_zone, color='red', linestyle='--', linewidth=1)

# --- Step 3: Add the text labels for each quadrant ---
# We place text by giving it x, y coordinates and the string.
# I've chosen coordinates that place the text in the corners of the quadrants.
plt.text(15, 82, 'Good Discipline', fontsize=12, color='green')
plt.text(44, 82, 'Aggressive', fontsize=12, color='black', ha='right')
plt.text(15, 53, 'Passive', fontsize=12, color='black')
plt.text(44, 53, 'Poor Discipline', fontsize=12, color='red', ha='right')

# --- Add Titles and Labels ---
plt.title('Player Swing Profiles (2025)', fontsize=16)
plt.xlabel('Out-of-Zone Swing % (O-Swing%)')
plt.ylabel('In-Zone Swing % (Z-Swing%)')

plt.grid(True)
plt.show()
```

```python
# --- Step 1: Import Plotly Express ---
import plotly.express as px

# --- Calculate the averages (we need these again) ---
avg_out_zone = df_merged['2025_out_zone'].mean()
avg_in_zone = df_merged['2025_in_zone'].mean()

# --- Step 2 & 3: Create the interactive scatter plot ---
fig = px.scatter(
    df_merged,
    x='2025_out_zone',
    y='2025_in_zone',
    hover_data=['name'], # To be able to hover cursor over datapoint and ID player name
    title='Interactive Player Swing Profiles (2025)',
    labels={ # A cleaner way to set labels in Plotly
        '2025_out_zone': 'Out-of-Zone Swing % (O-Swing%)',
        '2025_in_zone': 'In-Zone Swing % (Z-Swing%)'
    }
)

# --- Step 4: Add the quadrant lines ---
fig.add_vline(x=avg_out_zone, line_dash="dash", line_color="red")
fig.add_hline(y=avg_in_zone, line_dash="dash", line_color="red")

# --- Step 5: Add the quadrant labels using Plotly's method ---
# x= and y= give the pixel coordinates for the labels
fig.add_annotation(x=15, y=82, text="Good Discipline", showarrow=False, font=dict(color="green", size=14))
fig.add_annotation(x=46, y=82, text="Aggressive", showarrow=False, xanchor='right', font=dict(color="black", size=14))
fig.add_annotation(x=15, y=53, text="Passive", showarrow=False, font=dict(color="black", size=14))
fig.add_annotation(x=46, y=53, text="Poor Discipline", showarrow=False, xanchor='right', font=dict(color="red", size=14))

# Step 6. Fix the aspect ratio by setting a specific width and height
fig.update_layout(width=800, height=700)

# Step 7. Customize the hover text to show ONLY the player name
# %{customdata[0]} refers to the first column we passed into 'hover_data' (which was 'name')
# <extra></extra> is a special tag that removes the secondary box that sometimes appears
fig.update_traces(hovertemplate='<b>%{customdata[0]}</b><extra></extra>')


# --- Show the figure ---
fig.show()
```

```python
# --- Save the figure to an HTML file in order to share the interactive plot ---
fig.write_html("player_swing_profiles.html")
```

```python
# Create the new df_final by merging df_merged with a temporarily modified version of df_hitting
# This leaves BOTH df_merged and df_hitting in their original state.
df_final = pd.merge(
    df_merged,
    df_hitting.rename(columns={'player_id': 'entity_id'}), # Rename happens here, temporarily
    on='entity_id',
    how='inner'
)

# Let's look at the first few rows and the columns to confirm it worked
print("First 5 rows of the final merged data:")
print(df_final.head())
print("\nColumns in the final dataframe:")
print(df_final.columns)
```

```python
# --- Step 1: Import Plotly Express ---
import plotly.express as px

# --- Step 2: Recalculate the averages from our new df_final ---
# We use df_final now in case the inner merge removed any players
avg_out_zone = df_final['2025_out_zone'].mean()
avg_in_zone = df_final['2025_in_zone'].mean()


# --- Step 3: Create the interactive scatter plot ---
#    CHANGE 1: Use df_final as the data source
#    CHANGE 2: Add the 'color' argument to hue by xwoba
#    CHANGE 3: Add 'xwoba' to hover_data to see the value on hover
fig = px.scatter(
    df_final, # Change 1
    x='2025_out_zone',
    y='2025_in_zone',
    color='xwoba', # Change 2
    hover_data=['name', 'xwoba'], # Change 3
    title='Interactive Player Swing Profiles (2025) - Colored by xwOBA',
    labels={
        '2025_out_zone': 'Out-of-Zone Swing % (O-Swing%)',
        '2025_in_zone': 'In-Zone Swing % (Z-Swing%)',
        'xwoba': 'xwOBA' # Adds a clean label to the color bar
    }
)

# --- Step 4: Add the quadrant lines ---
fig.add_vline(x=avg_out_zone, line_dash="dash", line_color="red")
fig.add_hline(y=avg_in_zone, line_dash="dash", line_color="red")

# --- Step 5: Add the quadrant labels ---
fig.add_annotation(x=15, y=82, text="Good Discipline", showarrow=False, font=dict(color="green", size=14))
fig.add_annotation(x=46, y=82, text="Aggressive", showarrow=False, xanchor='right', font=dict(color="black", size=14))
fig.add_annotation(x=15, y=53, text="Passive", showarrow=False, font=dict(color="black", size=14))
fig.add_annotation(x=46, y=53, text="Poor Discipline", showarrow=False, xanchor='right', font=dict(color="red", size=14))

# --- Step 6: Fix the aspect ratio ---
fig.update_layout(width=900, height=800)

# --- Step 7: Customize the hover text ---
# The hovertemplate is a bit more complex now to show both name and xwoba cleanly
fig.update_traces(hovertemplate='<b>%{customdata[0]}</b><br>xwOBA: %{customdata[1]:.3f}<extra></extra>')


# --- Show the figure ---
fig.show()
```

```python
# Create the 'swing_differential' feature to see plate discipline as a single numerical column
df_final['swing_differential'] = df_final['2025_in_zone'] - df_final['2025_out_zone']

# Let's look at the result for a few players to see our new feature in action
print(df_final[['name', '2025_in_zone', '2025_out_zone', 'swing_differential', 'xwoba']].head())
```

```python
# Calculate the correlation between two specific columns
correlation_value = df_final['swing_differential'].corr(df_final['xwoba'])

print(f"The Pearson correlation between swing_differential and xwoba is: {correlation_value:.4f}")
```

```python
# The correlation value we found previously
correlation_value = 0.3696

# Calculate R-squared by squaring the correlation
r_squared = correlation_value ** 2

print(f"The R-squared value is: {r_squared:.4f}")
print(f"This can be interpreted as {r_squared:.1%} of the variance in xwOBA is explained by swing_differential.")
```

```python
# Create a list of all the columns we want to analyze
features_to_compare = [
    'xwoba',
    'swing_differential',
    'barrel_batted_rate',
    'hard_hit_percent',
    'blasts_swing',
    'ideal_angle_rate',
    'avg_best_speed'
]

# Select only these columns from our final dataframe
final_comparison_df = df_final[features_to_compare]

# Calculate the full correlation matrix
full_correlation_matrix = final_comparison_df.corr()

# To make the output easier to read, let's look at the correlations with 'xwoba' specifically,
# and sort them from highest to lowest.
xwoba_correlations = full_correlation_matrix['xwoba'].sort_values(ascending=False)

print("Correlations with xwOBA:")
print(xwoba_correlations)
```

```python
# Load the new expanded data into a DataFrame
df_hitting_expanded = pd.read_csv('2025_hitting_rate_stats_expanded.csv')

print("\nFirst 5 rows of the new expanded hitting data:")
print(df_hitting_expanded.head())
print("\nColumns in the expanded dataframe:")
print(df_hitting_expanded.columns)
```

```python
# Step 2: Merge the new expanded hitting data with our original swing data
# We use the same safe method as before, renaming the column just for the merge.
df_master = pd.merge(
    df_merged,
    df_hitting_expanded.rename(columns={'player_id': 'entity_id'}),
    on='entity_id',
    how='inner' # Ensures we only keep players that exist in both datasets
)

# Step 3: Re-create your 'swing_differential' feature in this new master DataFrame
df_master['swing_differential'] = df_master['2025_in_zone'] - df_master['2025_out_zone']

# Let's verify that the merge and feature creation worked
print("Merge and feature creation successful!")
print(f"The new df_master has {df_master.shape[0]} rows and {df_master.shape[1]} columns.")
print("\nHere are a few of the key columns to confirm:")
print(df_master[['name', '2025_in_zone', '2025_out_zone', 'swing_differential', 'bb_percent', 'avg_swing_speed']].head())
```

```python
# --- Define the skill and outcome metrics for our definitive analysis ---
features_for_final_analysis = [
    'swing_differential', # Your 'Discipline Skill'
    'avg_swing_speed',    # A key 'Physical Skill'
    'bb_percent',         # Proximal outcome of discipline
    'k_percent',          # Proximal outcome of discipline
    'barrel_batted_rate', # Proximal outcome of physical skill
    'hard_hit_percent',   # Proximal outcome of physical skill
    'exit_velocity_avg',  # Proximal outcome of physical skill
    'xwoba'               # The 'Final Outcome'
]

# Select only these columns from our master dataframe
master_comparison_df = df_master[features_for_final_analysis]

# Calculate the full correlation matrix
master_correlation_matrix = master_comparison_df.corr()


# --- Analysis Part 1: Your Primary Hypothesis ---
# How does the 'Discipline Skill' relate to its most direct outcomes?
discipline_correlations = master_correlation_matrix['swing_differential'].sort_values(ascending=False)

print("Correlations with swing_differential (The 'Discipline Skill'):")
print(discipline_correlations)
print("\n" + "="*60 + "\n") # Separator for clarity


# --- Analysis Part 2: Your Comparative Hypothesis ---
# How does a 'Physical Skill' relate to its most direct outcomes?
physical_skill_correlations = master_correlation_matrix['avg_swing_speed'].sort_values(ascending=False)

print("Correlations with avg_swing_speed (A 'Physical Skill'):")
print(physical_skill_correlations)
```

```python
# Calculate the averages for our quadrant lines
avg_discipline = df_master['swing_differential'].mean()
avg_speed = df_master['avg_swing_speed'].mean()

fig = px.scatter(
    df_master,
    x='swing_differential',
    y='avg_swing_speed',
    color='xwoba',
    hover_data=['name'],
    title='Player Archetypes: Discipline vs. Physical Skill',
    labels={
        'swing_differential': 'Discipline Skill (Swing Differential )',
        'avg_swing_speed': 'Physical Skill (Avg. Swing Speed )'
    }
)

# --- Add the quadrant lines ---
fig.add_vline(x=avg_discipline, line_dash="dash", line_color="gray")
fig.add_hline(y=avg_speed, line_dash="dash", line_color="gray")

# --- Corrected Quadrant Labels ---
fig.add_annotation(x=df_master['swing_differential'].min(), y=df_master['avg_swing_speed'].max(), text="<b>Aggressive Sluggers</b>", showarrow=False, xanchor='left', font=dict(color="black"))
fig.add_annotation(x=df_master['swing_differential'].max(), y=df_master['avg_swing_speed'].max(), text="<b>Unicorns</b>", showarrow=False, xanchor='right', font=dict(color="blue"))
fig.add_annotation(x=df_master['swing_differential'].min(), y=df_master['avg_swing_speed'].min(), text="<b>Aggressive Contact</b>", showarrow=False, xanchor='left', font=dict(color="red"))
fig.add_annotation(x=df_master['swing_differential'].max(), y=df_master['avg_swing_speed'].min(), text="<b>Patient / Selective</b>", showarrow=False, xanchor='right', font=dict(color="green"))


# --- Customize hover and layout ---
fig.update_traces(hovertemplate='<b>%{customdata[0]}</b><extra></extra>')

# --- THE FIX: Set a specific width and height for the figure ---
# You can adjust these pixel values to your liking.
fig.update_layout(
    width=1000,
    height=800
)

fig.show()
```
